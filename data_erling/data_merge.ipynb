```python
import pandas as pd
import numpy as np
```

```python
area_prices = pd.read_csv('area_prices.csv')
hourly_data = pd.read_csv('external_data/all_data_hourly.csv')

area_prices.drop(['Unnamed: 0', 'Unnamed: 0.1', 'Hours', 'datestring', 'SYS'], axis=1, inplace=True)

print(type(area_prices['datetime'].iloc[7]))

area_prices
```

```python
print(type(hourly_data['Date'].iloc[7]))
print(type(hourly_data['Hour'].iloc[7]))
hourly_data
```

```python
#Adding the area-prices to the hourly-data column

#Make %Y-%m-%d string for the dates and a column for the hour, then we match on datestring AND hour

from tqdm import tqdm

area_prices['Hour'] = None
area_prices['Date'] = None

for row in tqdm(range(len(area_prices))):
    area_prices['Date'].iloc[row] = area_prices['datetime'].iloc[row][:10]
    area_prices['Hour'].iloc[row] = np.int64(area_prices['datetime'].iloc[row][11:13])

area_prices
```

```python
#Making datetime of the different datetime columns

area_prices['Date'] = pd.to_datetime(area_prices['Date'])
hourly_data['Date'] = pd.to_datetime(hourly_data['Date'])

min_date = hourly_data['Date'].iloc[0]
max_date = hourly_data['Date'].iloc[-1]

print(min_date)
print(max_date)

rows = []

for row in tqdm(range(len(area_prices))):
    if area_prices['Date'].iloc[row] < min_date or area_prices['Date'].iloc[row] > max_date:
        rows.append(row)

area_prices.drop(rows, inplace=True, axis=0)
area_prices
```

```python
#finding columns which are excesssive or of non are printed, there are likely duplicates
#We are supposed to have 61368 rows of data

area_prices.drop_duplicates(subset=['Date', 'Hour'], inplace=True)

area_prices
```

```python
area_columns = ['SE1', 'SE2', 'SE3', 'SE4', 'FI', 'DK1', 'DK2', 'Oslo', 'Kr.sand',
       'Bergen', 'Molde', 'Tr.heim', 'Tromsø']

hourly_data
```

```python
area_prices.reset_index(inplace=True)
```

```python
for col in area_columns:
    hourly_data[col] = area_prices[col].astype(np.float64)
    print(hourly_data.shape)

hourly_data.columns
```

```python
#Reading the daily data

#Columns which are seperate to the daily data
daily_cols_unincluded = ['Temp Hamar', 'Temp Krsand', 'Temp Namsos', 'Temp Troms',
              'Temp Bergen', 'Temp Norway', 'Supply', 'Demand', 'Oil',
              'Coal', 'Gas', 'Low Carbon', 'APX', 'OMEL', 'EEX', 'Wind SE',
              'Wind Prod', 'Prec Norway', 'Prec Norway 7', 'Snow Norway', 'Date']

#Reading the daily prices and creating a dataFrame
daily_prices = pd.read_csv('external_data/all_data_daily.csv')
daily_prices['Date'] = pd.to_datetime(daily_prices['Date'])
daily_cols_unincluded.remove('Date')
hourly_data[daily_cols_unincluded] = None
#daily_cols_unincluded.append('Date')


#Creating a numpy array to represent the granularity of the data, which is basically to be copied thorughout the day
time_series_numpy = hourly_data[['Date', 'Oil', 'Hour']]
time_series_numpy.columns =['Date', 'Value', 'Hour']
time_series_numpy = time_series_numpy.to_numpy()


#Iterating over the columns, and for each column mapping this into the hourly data dataFrame
for col in tqdm(daily_cols_unincluded):
    idx_day = 0
    daily_prices_numpy = daily_prices[['Date', col]].to_numpy()
    for row in range(len(time_series_numpy)):
        if time_series_numpy[row, 0] != daily_prices_numpy[idx_day, 0]:
            print('ERROR')
        else:
            time_series_numpy[row, 1] = daily_prices_numpy[idx_day, 1]
        if time_series_numpy[row, 2] == 23:
            idx_day += 1
    hourly_data[col] = time_series_numpy[:, 1]
    
    

#daily_prices_numpy = daily_prices.to_numpy()
#dates_numpy = hourly_data[daily_cols_unincluded].to_numpy()

hourly_data[daily_cols_unincluded].max()
```

```python
#Saving as .csv

hourly_data.to_csv('hourly_data_areas.csv')
```
